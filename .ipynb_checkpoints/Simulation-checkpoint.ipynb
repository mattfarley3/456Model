```python
import numpy as np
import pandas as pd
import plotly.express as px
import random as rd
import json
import time
```

```python
class Simulation():
    def __init__(self, n, t):
        self.t = t
        self.population = n
        self.x = np.random.rand(self.population)*10
        self.y = np.random.rand(self.population)*10
        data = {'x_value': self.x, 'y_value': self.y, 'id': np.arange(n), 'frame':np.zeros(n)}
        self.df = pd.DataFrame(data)
        self.step()
        px.scatter(self.df, x="x_value", y="y_value", animation_frame="frame", animation_group="id", color="id", range_x=[0,10], range_y=[0,10])
        
        
    def step(self):
        for i in range(self.t):
            for j in range(self.population):
                x = rd.randint(0,2)
                y = rd.randint(0,2)
                if x == 1:
                    if self.x[j]+0.5<=10:
                        self.x[j]+=0.5
                elif x == 2:
                    if self.x[j]-0.5>=0:
                        self.x[j]-=0.5
                if y == 1:
                    if self.y[j]+0.5<=10:
                        self.y[j]+=0.5
                elif y == 2:
                    if self.y[j]-0.5>=0:
                        self.y[j]-=0.5
            data = {'x_value': self.x, 'y_value': self.y, 'id': np.arange(self.population), 'frame': np.ones(self.population)*i}
            self.df = pd.concat([self.df, pd.DataFrame(data)])
            
```

```python
s = Simulation(50, 50)
```

```python
px.scatter(s.df, x="x_value", y="y_value", animation_frame="frame", animation_group="id", color="id", range_x=[0,10], range_y=[0,10])
```
